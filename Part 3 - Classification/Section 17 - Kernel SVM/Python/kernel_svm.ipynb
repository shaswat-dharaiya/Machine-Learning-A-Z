# Kernel SVM

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataSet = pd.read_csv('Social_Network_Ads.csv')
data_X = dataSet.iloc[:, :-1].values
data_Y = dataSet.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_Y, test_Y = train_test_split(data_X, data_Y, test_size=0.2, random_state=0)
```

```python
print(train_X)
```

```python
print(train_Y)
```

```python
print(test_X)
```

```python
print(test_Y)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
train_X = sc.fit_transform(train_X)
test_X = sc.transform(test_X)
```

```python
print(train_X)
```

```python
print(test_X)
```

## Training the Kernel SVM model on the Training set

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(train_X, train_Y)
```

## Predicting a new result

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

## Predicting the Test set results

```python
pred_Y = classifier.predict(test_X)
print(np.concatenate((pred_Y.reshape(len(pred_Y),1), test_Y.reshape(len(test_Y),1)),1))
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(test_Y, pred_Y)
print(cm)
accuracy_score(test_Y, pred_Y)
```

## Visualising the Training set results

```python
from matplotlib.colors import ListedColormap
set_X, set_Y = sc.inverse_transform(train_X), train_Y
X1, X2 = np.meshgrid(np.arange(start = set_X[:, 0].min() - 10, stop = set_X[:, 0].max() + 10, step = 0.25),
                     np.arange(start = set_X[:, 1].min() - 1000, stop = set_X[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(set_Y)):
    plt.scatter(set_X[set_Y == j, 0], set_X[set_Y == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Kernel SVM (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

## Visualising the Test set results

```python
from matplotlib.colors import ListedColormap
set_X, set_Y = sc.inverse_transform(test_X), test_Y
X1, X2 = np.meshgrid(np.arange(start = set_X[:, 0].min() - 10, stop = set_X[:, 0].max() + 10, step = 0.25),
                     np.arange(start = set_X[:, 1].min() - 1000, stop = set_X[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(set_Y)):
    plt.scatter(set_X[set_Y == j, 0], set_X[set_Y == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Kernel SVM (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
